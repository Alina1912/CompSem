```python
file = "model.bin"
model = gensim.models.KeyedVectors.load_word2vec_format(file, binary=True)
```

```python
words = []
file = open("HW2_igrat'.txt", 'r', encoding = 'utf-8') 
for line in file:
    a = line.split()[0]
    b = a.split("_")[1]    
    words.append(b)
print(words)
```

```python
words_play = [] 
file = open("HW2_igrat'.txt", 'r', encoding = 'utf-8') 
for line in file:
    a = line.split()[0]
    words_play.append(a)
print(words_play)
```

```python
new_words = []
for word in words:
    new_words.append(word + "_NOUN")
print(new_words)
```

```python
for word in new_words:
    # есть ли слово в модели? 
    if word in model:
        print(word)
        # смотрим на вектор слова (его размерность 300, смотрим на первые 10 чисел)
        print(model[word])
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy
import scipy.cluster.hierarchy as hcluster
```

```python
dmfile = 'smooth_ruscorp_wa_ppmi_300_norm.dm'
dictionary = {}
with open('smooth_ruscorp_wa_ppmi_300_norm.dm','w') as f:
    for word in new_words:
            vec = model[word] + model['играть_VERB']
            dictionary[f'играть_{word[:-5]}'] = vec
            #print(vec)
            array = list(map(str, vec))
            l = '\t'.join(array)
            f.write(f'играть_{word[:-5]}\t{l}\n')
```

```python
array = []
with open (dmfile, 'r') as f_input:
    for line in f_input:
        line = line.strip().split('\t')
        vector = []
        for item in line[1:]:
            vector.append(float(item))
        array.append(vector)
```

```python
X = np.array(array[:13])
print (X.shape)
```

```python
Z = hcluster.linkage(X)
plt.figure()
dn = hcluster.dendrogram(Z)
plt.show()
```

```python
clusters = hcluster.fcluster(Z, 1)
print('Estimated number of clusters: %d' % len(set(clusters)))
```

```python
clusters = hcluster.fcluster(Z, 1.15)
print('Estimated number of clusters: %d' % len(set(clusters)))
```

```python
clusters = hcluster.fcluster(Z, 0.8)
print('Estimated number of clusters: %d' % len(set(clusters)))
```

```python

clusters = hcluster.fclusterdata(X, 1)
print('Estimated number of clusters: %d' % len(set(clusters)))
```
